# Train Test Split

```python
import pandas as pd
```

```python
test = pd.read_csv('test_v3.0.csv')
```

```python
train= pd.read_csv('train_v3.0.csv')
```

```python
train['code'] = train.iloc[:,-3:-2].apply(lambda x: pd.factorize(x)[0])
train['label']=train['font'].map(str)+' '+train['code'].map(str)
```

```python
test['code'] = test.iloc[:,-3:-2].apply(lambda x: pd.factorize(x)[0])
test['label']=test['font'].map(str)+' '+test['code'].map(str)
```

```python
train
```

```python
test['font'].value_counts()
```

```python
train['font'].value_counts()
```

```python
mix=train.append(test)
```

```python
mix['font'].value_counts()
```

```python
mix=mix.iloc[:, 2:7]
```

```python
mix
```

```python
from sklearn.utils import shuffle


mix_shuffled = shuffle(mix, random_state=100)
```

```python
mix_shuffled=mix_shuffled.reset_index()
```

```python
mix_shuffled
```

```python
test_number=(mix['label'].value_counts()*0.3).to_dict()
```

```python
mix['label'].value_counts()
```

```python
train_number=(mix['label'].value_counts()*0.7).to_dict()
```

```python
train_number
```

```python
test_number
```

```python
count_number=test_number.copy()
```

```python
count_number
```

```python
count_number = dict.fromkeys(count_number, 0)
```

```python
test_number['2 1888']
```

```python
test_list=[]
train_list=[]
```

test would have more data

```python
for index, row in mix_shuffled.iterrows():
    if count_number[row.label]<test_number[row.label]:
        count_number[row.label]=count_number[row.label] + 1
        test_list.append([row.path,row.font,row.char,row.code,row.label])
    else:
        train_list.append([row.path,row.font,row.char,row.code,row.label])
```

```python
train_df = pd.DataFrame(train_list, columns=['path','font','char','code','label'])
test_df = pd.DataFrame(test_list, columns=['path','font','char','code','label'])
```

```python
train_df['font'].value_counts()
```

```python
test_df['font'].value_counts()
```

```python
train_df.to_csv('train_small_4.0.csv', index=False)
```

```python
test_df.to_csv('test_small_4.0.csv', index=False)
```

train more data

```python
count_number = dict.fromkeys(count_number, 0)
```

```python
test_list_2=[]
train_list_2=[]
```

```python
train_df_2 = pd.DataFrame(train_list_2, columns=['path','font','char','code','label'])
test_df_2 = pd.DataFrame(test_list_2, columns=['path','font','char','code','label'])
```
